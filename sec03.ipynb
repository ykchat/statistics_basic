```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import seaborn as sns

sns.set()
```

#### 27. 確率関数とパラメータ

```python
p = 0.3 #パラメータ

x = np.arange(0, 2)
y = stats.bernoulli.pmf(k=x, p=p)

df = pd.DataFrame({"x": x, "probability": y})
sns.barplot(x="x", y="probability", data=df)
```

#### 29. 二項分布

```python
n = 4
p = 0.3

x = np.arange(0, n+1)
y = stats.binom.pmf(k=x, n=n, p=p)

df = pd.DataFrame({"x": x, "probability": y})
sns.barplot(x="x", y="probability", data=df)
```

#### 31. 正規分布

```python
mu = 170 #平均
sigma = 5 #標準偏差

x = np.linspace(start=150, stop=190, num=101)
y = stats.norm.pdf(x=x, loc=mu, scale=sigma)

df = pd.DataFrame({"x": x, "probability": y})
sns.lineplot(x="x", y="probability", data=df)
```

#### 32. 標準正規分布

```python
mu = 0 #平均
sigma = 1 #標準偏差

x = np.linspace(start=-4, stop=4, num=101)
y = stats.norm.pdf(x=x, loc=mu, scale=sigma)

df = pd.DataFrame({"x": x, "probability": y})
sns.lineplot(x="x", y="probability", data=df)
```
