```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import seaborn as sns

sns.set()
```

#### 27. 確率関数とパラメータ

```python
p = 0.3 #パラメータ

x = np.arange(0, 2)
df = pd.DataFrame(index=x)

y = stats.bernoulli.pmf(k=x, p=p)
df = pd.concat([df, pd.DataFrame({"probability": y}, index=x)], axis=1)

sns.barplot(x=df.index, y=df.probability, data=df)
```

#### 29. 二項分布

```python
n = 4
p = 0.3

x = np.arange(0, n+1)
df = pd.DataFrame(index=x)

y = stats.binom.pmf(k=x, n=n, p=p)
df = pd.concat([df, pd.DataFrame({"probability": y}, index=x)], axis=1)

sns.barplot(x=df.index, y=df.probability, data=df)
```

#### 31. 正規分布

```python
mu = 170 #平均
sigma = 5 #標準偏差

x = np.linspace(
    start=stats.norm.ppf(q=0.0001, loc=mu, scale=sigma), 
    stop=stats.norm.ppf(q=0.9999, loc=mu, scale=sigma), 
    num=101)
df = pd.DataFrame(index=x)

y = stats.norm.pdf(x=x, loc=mu, scale=sigma)
df = pd.concat([df, pd.DataFrame({"probability": y}, index=x)], axis=1)

sns.lineplot(x=df.index, y=df.probability, data=df)
```

#### 32. 標準正規分布

```python
mu = 0 #平均
sigma = 1 #標準偏差

x = np.linspace(
    start=stats.norm.ppf(q=0.0001, loc=mu, scale=sigma), 
    stop=stats.norm.ppf(q=0.9999, loc=mu, scale=sigma), 
    num=101)
df = pd.DataFrame(index=x)

y = stats.norm.pdf(x=x, loc=mu, scale=sigma)
df = pd.concat([df, pd.DataFrame({"probability": y}, index=x)], axis=1)

sns.lineplot(x=df.index, y=df.probability, data=df)
```
